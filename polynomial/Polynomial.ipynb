```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
years = np.array([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]).reshape(-1, 1)
prices = np.array([400, 600, 1000, 3700, 7200, 29000, 47000, 32000, 46000, 58000])
```

```python
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(years)
```

```python
model = LinearRegression()
model.fit(X_poly, prices)
```

```python
year_input = int(input("Bitcoin narxini bashorat qilish uchun yilni kiriting: "))
predicted_price = model.predict(poly.transform([[year_input]]))
print(f"{year_input}-yilda Bitcoin narxi taxminan {predicted_price[0]:.2f} $ bo‘lishi mumkin.")
```

```python
future_years = np.linspace(2015, 2030, 100).reshape(-1, 1)
future_prices = model.predict(poly.transform(future_years))

plt.scatter(years, prices, color='red', label="Real ma'lumotlar")
plt.plot(future_years, future_prices, color='blue', label="Polynomial Regression bashorati")
plt.scatter(year_input, predicted_price, color='green', marker='x', s=100, label="Bashorat qilingan narx")
plt.xlabel("Yillar")
plt.ylabel("Bitcoin narxi ($)")
plt.legend()
plt.show()
```
